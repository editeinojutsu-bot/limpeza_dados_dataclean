```python
import pandas as pd
df = pd.read_csv("/content/dados_clientes.csv")
df.head(5)
```

```python
df["data_compra"] = pd.to_datetime(df["data_compra"], errors='coerce')
print(df.info())
print(df.describe())
print(df.isna().sum())
print(df.duplicated().sum())
```

```python
print("Número total de registro são", df["id_cliente"].count())

# Tratando valores ausentes

# por ter mais de 30% de valores ausentes, vamos excluir a coluna data_compra
df = df.drop('data_compra', axis=1)
print(df.isna().sum())

# Para preencher os valores ausentes, na idade e na renda_anual vamos preencher com a média
# e no valor da compra vamos preencher com a média da categoria_produto
df['idade'] = df['idade'].fillna(df['idade'].mean())
df['renda_anual'] = df['renda_anual'].fillna(df['renda_anual'].mean())
df['valor_compra'] = df.groupby('categoria_produto')['valor_compra'].transform(lambda x: x.fillna(x.mean()))
print("Valores ausentes após preenchimento do valor_compra:")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# análise de outliers
numerical_cols = ['idade', 'renda_anual', 'valor_compra']

plt.figure(figsize=(15, 5))
for i, col in enumerate(numerical_cols):
    plt.subplot(1, len(numerical_cols), i + 1)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot de {col}')
    plt.ylabel('')
plt.tight_layout()
plt.show()
```

```python
def find_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    return outliers

numerical_cols = ['idade', 'renda_anual', 'valor_compra']

print("Outliers identificados pelo método IQR:")
for col in numerical_cols:
    outliers = find_outliers_iqr(df, col)
    if not outliers.empty:
        print(f"\nColuna '{col}':")
        display(outliers[[col]])
    else:
        print(f"\nColuna '{col}': Nenhum outlier encontrado.")

print("Por ser uma amostra pequeno, vamos manter esses ourtliers, pq possivelmente não são outliers")
```

```python
from sklearn.preprocessing import MinMaxScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
import numpy as np

# Definir colunas numéricas e categóricas
numerical_cols = ['idade', 'renda_anual', 'valor_compra']
categorical_cols = ['genero', 'categoria_produto', 'regiao']

# Criar um ColumnTransformer para aplicar diferentes transformações a diferentes colunas
preprocessor = ColumnTransformer(
    transformers=[
        ('num', MinMaxScaler(), numerical_cols),
        ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_cols)
    ])

# Aplicar as transformações
df_transformed = preprocessor.fit_transform(df)

# Obter nomes das novas colunas para o OneHotEncoder
cat_feature_names = preprocessor.named_transformers_['cat'].get_feature_names_out(categorical_cols)

# Concatenar nomes das colunas numéricas e categóricas transformadas
new_column_names = numerical_cols + list(cat_feature_names)

# Criar um novo DataFrame com os dados transformados
df_processed = pd.DataFrame(df_transformed, columns=new_column_names)

print("DataFrame após normalização e codificação (primeiras 5 linhas):")
display(df_processed.head())
```
